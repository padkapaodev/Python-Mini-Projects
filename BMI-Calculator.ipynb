# BMI Calculator
**Formula:** BMI = weight (kg) / (height (m))²
* **Severe Thinness:** < 16
* **Moderate Thinness:** 16 - 17
* **Mild Thinness:** 17 - 18.5
* **Normal:** 18.5 - 25
* **Overweight:** 25 - 30
* **Obese Class I:** 30 - 35
* **Morbidly Obese:** > 35

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Name = input("Enter your name: ")
Weight = int(input("Enter your weight in kg: "))
Height = int(input("Enter your height in cm: "))
BMI = Weight / ((Height/100)**2)  # Change height from cm to m by dividing by 100

print("Your name is", Name)
print("Your BMI is", BMI)

if BMI > 0:
    if BMI < 16:
        print(Name + ", You are Severe Thinness")
    elif BMI < 18.5:
        print(Name + ", You are Mild Thinness")
    elif BMI <= 25:
        print(Name + ", You are Normal")
    elif BMI < 30:
        print(Name + ", You are Overweight")
    elif BMI >= 30 and BMI < 35 :
        print(Name + ", You are Obese")
    else:
        print(" You are Morbidly Obese")
else:
    print("Enter invalid input")

x_user = Name
y_bmi = BMI

fig, ax = plt.subplots(figsize=(6, 8))

# Fill in the zone with color (axhspan)
ax.axhspan(0, 16, color='#e74c3c', alpha=0.3, label='Severe Thinness (< 16)')
ax.axhspan(16, 18.5, color='#e67e22', alpha=0.3, label='Mild/Moderate Thinness (16 - 18.5)')
ax.axhspan(18.5, 25, color="#26cf6d", alpha=0.3, label='Normal (18.5 - 25)')
ax.axhspan(25, 30, color='#f1c40f', alpha=0.3, label='Overweight (25 - 30)')
ax.axhspan(30, 35, color='#e67e22', alpha=0.3, label='Obese Class I (30 - 35)')
ax.axhspan(35, 50, color='#e74c3c', alpha=0.3, label='Morbidly Obese (> 35)')

# Plot the user's BMI points
ax.scatter(x_user, y_bmi, color='red', s=250, zorder=5, edgecolor='black', linewidth=1.5)

# Label the BMI number on the point (rounded to 2 decimal places)
ax.annotate(f'BMI: {y_bmi:.2f}', (x_user, y_bmi), fontsize=12, weight='bold', 
            va='bottom', ha='center', xytext=(0, 10), textcoords='offset points')

# set the axes and the appearance of the graph.
ax.set_ylim(10, 40)  # ปรับช่วงของแกน Y ให้ครอบคลุมทุกเกณฑ์
ax.set_ylabel('BMI Value', fontsize=12, weight='bold')
ax.set_xlabel('User', fontsize=12, weight='bold')
plt.title(f"{Name}'s BMI Report", fontsize=14, weight='bold', pad=15)

plt.legend(bbox_to_anchor=(1.05, 1), title="BMI Categories", title_fontproperties={'weight':'bold'})
plt.tight_layout()
plt.show()
```
