```python
import os,shutil
```

```python
path = r"C:/Users/Padka/OneDrive/Desktop/Learning Python/Review overview/" #raw str
```

```python
file_name = os.listdir(path) #check files in the path
```

```python
folder_name = ['Goku', 'Green', 'Zoro']

# create folders if they do not exist
for loop in range(len(folder_name)):
    if not os.path.exists(path + folder_name[loop]):
        os.makedirs(path + folder_name[loop])

# move files to the corresponding folders based on their names
for file in file_name:
    if 'Goku.jpg' in file and not os.path.exists(path + 'Goku/' + file):
        shutil.move(path + file, path + 'Goku/' + file)
    if 'Green.jpg' in file and not os.path.exists(path + 'Green/' + file):
        shutil.move(path + file, path + 'Green/' + file)
    if 'Zoro.png' in file and not os.path.exists(path + 'Zoro/' + file):
        shutil.move(path + file, path + 'Zoro/' + file)
    else:
        print("There are files in this path that were not moved")
```
